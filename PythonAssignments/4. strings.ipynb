```python
# Assignment 1: Create a string that contains a simple bio data and extract information
bio_data = "Name: Alice, Age: 30, Country: USA"
name = bio_data.split(", ")[0].split(": ")[1]
age = bio_data.split(", ")[1].split(": ")[1]
print("Extracted Info:", f"Name: {name}, Age: {age}")
```

```python
# Assignment 2: Format string using a dictionary
info = {"name": "Bob", "age": 25, "country": "Canada"}
format = f"{info['name']} is {info['age']} years old and lives in {info['country']}."
print("Formatted Info:", format)
```

```python
# Assignment 3: Character count in a string
def char_count(string):
    return {char: string.count(char) for char in set(string)}

string = "Hello, world!"
print("Character Count:", char_count(string))
```

```python
# Assignment 4
import re
phone_regex = r"\b(?:\+880|0088)?1[3-9]\d{8}\b|\b1[3-9]\d{2}-\d{6}\b"
phones_text = "Valid numbers: 01454565767, +880196345634, 0088785674657, 01845-567567."
bd_numbers = re.findall(phone_regex, phones_text)
print("Bangladesh Phone", bd_numbers)
```
