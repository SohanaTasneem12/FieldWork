```python
# Assignment 1: Write a function that calculates compound interest using the formula A = P(1 + r/n)^(nt).
def calculate_compound_interest(P, r, n, t):
    A = P * (1 + r / n) ** (n * t)
    return A

principal = 1000
rate = 0.05
compounds_per_year = 4
time = 5

compound_amount = calculate_compound_interest(principal, rate, compounds_per_year, time)
print(f"Compound interest result: {compound_amount:.2f}")
```

```python
# Assignment 2: Create a script that prints the current time and updates every second.
import datetime

def print_current_time():
    while True:
        current_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print("Current time:", current_time, end='\r')
        time.sleep(1)
print_current_time()
```
